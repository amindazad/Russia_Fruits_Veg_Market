```
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
```

## Total Fruits & Vegetables Imports 2010 - 2019


```
rus_fv_total = pd.read_csv("https://raw.githubusercontent.com/amindazad/Russia_Fruits_Veg_Market/main/Datasets/Total%20Fruits%20and%20Vegetables%20Imports/07%2B08%201996-2019%20imports.csv").sort_values(by="Year")

rus_fv_total = rus_fv_total.drop([ 'Period', 'Period Desc.', 'Trade Flow', 
                       '2nd Partner Code', '2nd Partner','2nd Partner ISO',
                       'Customs Proc. Code','Customs','Mode of Transport Code',
                       'Mode of Transport','Alt Qty Unit Code','Alt Qty Unit','Alt Qty',
                        'Gross weight (kg)','CIF Trade Value (US$)','FOB Trade Value (US$)'],axis=1)

rus_fv_total = rus_fv_total.replace({'Edible vegetables and certain roots and tubers':'Vegetables',
                                     'Edible fruit and nuts; peel of citrus fruit or melons' : 'Fruits',
                                     'Vegetables and certain roots and tubers; edible':'Vegetables',
                                     'Fruit and nuts, edible; peel of citrus fruit or melons' : 'Fruits', 
                 "Edible fruit, nuts, peel of citrus fruit, melons":"Fruits"})
rus_fv_total
```

```
rus_fv_total_fig_1 = px.bar(rus_fv_total, x="Year", y="Trade Value (US$)",
                            color="Commodity",barmode='overlay',
                            template="plotly_white",text="Trade Value (US$)", 
                            title="Russia Fruits & Vegetables Import 1996-2019 (Source: UN Comtrade 2019)")

rus_fv_total_fig_1.update_traces(texttemplate='%{text:.3s}', 
                                 textposition='outside')
#rus_fv_total_fig_1.for_each_annotation(lambda a: a.update(text=a.text.split("=")[1]))
rus_fv_total_fig_1.for_each_trace(lambda t: t.update
                                  (name=t.name.replace("Commodity=", "")))

rus_fv_total_fig_1.show()
```

## Fruits & Vegtetables by Category 2010-2019

```
fg_2010_2019_bycategory = pd.concat([veg_2010_2019_bycategory, fruits_bycategory_2010_2019]).sort_values(by="Year").reset_index()

fg_2010_2019_bycategory.head()
```

```
fg_2010_2019_bycategory['Average_Price'] = fg_2010_2019_bycategory['Trade Value (US$)'] /fg_2010_2019_bycategory['Netweight (kg)']

fg_2010_2019_bycategory = fg_2010_2019_bycategory.sort_values(by='Average_Price', ascending=False)

fg_2010_2019_bycategory.head()
```

```
fg_2010_2019_bycategory_fig = px.bar(fg_2010_2019_bycategory, x="Year", 
                               y="Trade Value (US$)", 
                               color='Commodity',
                               template='plotly_white',
                               title="Russia Fruits & Vegetables Imports by Category (Source: UN Comtrade 2019)",
                            barmode='overlay')

fg_2010_2019_bycategory_fig.for_each_trace(lambda t: t.update
                                  (name=t.name.replace("Commodity=", "")))

fg_2010_2019_bycategory_fig.show()
```
