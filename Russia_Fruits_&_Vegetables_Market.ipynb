```
import pandas as pd
import numpy as np
import plotly.express as px
```

## Total Fruits & Vegetables Imports 2010 - 2019


```
fv_2010_2014 = pd.read_csv("https://raw.githubusercontent.com/amindazad/Russia_Fruits_Veg_Market/main/f%26g_2010-2015.csv")
fv_2015_2019 = pd.read_csv("https://raw.githubusercontent.com/amindazad/Russia_Fruits_Veg_Market/main/f%26g_2015-2019.csv")
fv_2010_2019 = pd.concat([fv_2015_2019,fv_2010_2014]).sort_values(by=['Year']).reset_index()
fv_2010_2019 = fv_2010_2019.drop(['index', 'Period', 'Period Desc.', 'Trade Flow', 
                       '2nd Partner Code', '2nd Partner','2nd Partner ISO',
                       'Customs Proc. Code','Customs','Mode of Transport Code',
                       'Mode of Transport','Alt Qty Unit Code','Alt Qty Unit','Alt Qty',
                       'Gross weight (kg)','CIF Trade Value (US$)','FOB Trade Value (US$)'],axis=1)
fv_2010_2019 = fv_2010_2019.replace({'Edible vegetables and certain roots and tubers':'Vegetables',
                                     'Edible fruit and nuts; peel of citrus fruit or melons' : 'Fruits',
                                     'Vegetables and certain roots and tubers; edible':'Vegetables',
                                     'Fruit and nuts, edible; peel of citrus fruit or melons' : 'Fruits',})
fv_2010_2019
```

```
fv_2010_2019_fig_1 = px.bar(fv_2010_2019, x="Year", y="Trade Value (US$)",color="Commodity")
fv_2010_2019_fig_1.show()
```

## Ftuit imports by category 2010 - 2019

```
fruits_byitem_2010_2014 = pd.read_csv('https://raw.githubusercontent.com/amindazad/Russia_Fruits_Veg_Market/main/2010-2014%20Fruits%20import%20by%20item.csv')
fruits_byitem_2015_2019 = pd.read_csv('https://raw.githubusercontent.com/amindazad/Russia_Fruits_Veg_Market/main/2015-2019%20Fruits%20import%20by%20item.csv')
fruits_byitem_2010_2019 = pd.concat([fruits_byitem_2010_2014, fruits_byitem_2015_2019]).reset_index().sort_values(by=['Year','Trade Value (US$)'], ascending=False)
fruits_byitem_2010_2019 = fruits_byitem_2010_2019.drop(['index', 'Period', 'Period Desc.', 'Trade Flow', 
                       '2nd Partner Code', '2nd Partner','2nd Partner ISO',
                       'Customs Proc. Code','Customs','Mode of Transport Code',
                       'Mode of Transport','Alt Qty Unit Code','Alt Qty Unit','Alt Qty',
                       'Gross weight (kg)','CIF Trade Value (US$)','FOB Trade Value (US$)'],axis=1)
fruits_byitem_2010_2019 = fruits_byitem_2010_2019.replace({'Citrus fruit, fresh or dried.':'Citrus fruit; fresh or dried',
                                     'Bananas, including plantains, fresh or dried.' : 'Bananas, including plantains; fresh or dried',
                                     'Apples, pears and quinces, fresh.':'Apples, pears and quinces; fresh',
                                     'Apricots, cherries, peaches (including nectarines), plums and sloes, fresh.' : 'Apricots, cherries, peaches (including nectarines), plums and sloes, fresh',
                                     'Grapes, fresh or dried.':'Grapes; fresh or dried',
                                     'Dates, figs, pineapples, avocados, guavas, mangoes and mangosteens, fresh or dried.':'Dates, figs, pineapples, avocados, guavas, mangoes and mangosteens; fresh or dried',
                                     'Other fruit, fresh.':'Fruit, fresh; n.e.c. in chapter 08',
                                     'Fruit and nuts, uncooked or cooked by steaming or boiling in water, frozen, whether or not containing added sugar or other sweetening matter.':'Fruit and nuts; uncooked or cooked by steaming or boiling in water, frozen, whether or not containing added sugar or other sweetening matter',
                                     'Coconuts, Brazil nuts and cashew nuts, fresh or dried, whether or not shelled or peeled.':'Nuts, edible; coconuts, Brazil nuts and cashew nuts, fresh or dried, whether or not shelled or peeled',
                                     'Nuts (excluding coconuts, Brazils and cashew nuts); fresh or dried, whether or not shelled or peeled':'Other nuts, fresh or dried, whether or not shelled or peeled.',
                                     'Melons (including watermelons) and papaws (papayas), fresh.':'Melons (including watermelons) and papaws (papayas); fresh',
                                     'Fruit and nuts, provisionally preserved (for example, by sulphur dioxide gas, in brine, in sulphur water or in other preservative solutions), but unsuitable in that state for immediate consumption.':'Fruit and nuts provisionally preserved; e.g. by sulphur dioxide gas, brine, in sulphur water or in other preservative solutions, but unsuitable in that state for immediate consumption',
                                     'Fruit, dried, other than that of headings 08.01 to 08.06; mixtures of nuts or dried fruits of this Chapter.':'Fruit, dried, other than that of heading no. 0801 to 0806; mixtures of nuts or dried fruits of this chapter',
                                     'Peel of citrus fruit or melons (including watermelons), fresh, frozen, dried or provisionally preserved in brine, in sulphur water or in other preservative solutions.':'Peel of citrus fruit or melons (including watermelons); fresh, frozen dried or provisionally preserved in brine, in sulphur water or in other preservative solutions'})

pd.set_option("display.max_colwidth", None)
pd.set_option("display.max_rows", None)
fruits_byitem_2010_2019
```

```
fig = px.bar(fruits_byitem_2010_2019, x="Year", y="Trade Value (US$)", color='Commodity')
fig.show()
```

```
fig_2 = px.pie(fruits_byitem_2010_2019, values='Trade Value (US$)', names='Commodity')
fig_2.show()
```
